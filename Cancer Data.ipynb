# Split the Data

```python
x = df.iloc[:,2:31].values
y = df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

# Decision Tree
